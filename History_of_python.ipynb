#History of python and some of its features 

The history of the Python programming language dates back to the late 80s and early 90s1, its implementation began in December 19892 when at Christmas Guido Van Rossum who worked at the (CWI) (an official Dutch research center that , among other things, it currently houses the W3C central office) decided to start the project as a hobby, continuing the ABC programming language of which he had been part of the development team at the CWI, this language focused on being easy to use and learn while maintaining power in its performance but the hardware available at the time of its creation made it difficult to use and the project did not transcend as expected. Van Rossum is therefore the main author of Python and continues to play a central role in deciding the direction of the language, in the Python community he is known as Benevolent Dictator for Life (BDFL). The name "Python" is given by Van Rossum's fondness for the Monty Python group.

The Python programming language was originally developed for the Amoeba operating system. 




Van Rossum wrote in 1996: 

"...Six years ago, in December 1989, I was looking for a hobby programming project that would keep me busy during the Christmas weeks. My office would be closed and I should have nothing but my home computer on hand. I decided to write an interpreter for the new scripting language I had recently been coming up with: a descendant of ABC that Unix / c hackers would like. I chose the Python name for the project, finding myself in a slightly irreverent frame of mind (and being a huge fan of Monty Python’s Flying Circus). ” 


In 2000 he added:

"Python's predecessor, ABC, was inspired by SETL - Lambert Meertens spent a year working in the SETL group at New York University before the final design for ABC emerged.6 " 

With these words, it can be interpreted that Rossum was unaware of the scope of its creation, it was directed for a group of people and is now universal in scope, has various criticisms and things to improve but in general its virtues are greater and it is advantageous for possess a high-level interpreted language that can be extended with C or C ++, It has several programming environments that allow editing programs, interacting with the interpreter, developing projects, debugging, among others, and is currently supported by a large community that facilitates learning and produces progress in new versions. 

This multiparadigm language called Python offers tools to work from object-oriented programming, as well as functional programming from list and imperative programming, different types of applications can be programmed: scripts, web applications and desktop applications. It has hundreds of libraries that make any type of project possible, be it a mobile application, web application, data science or artificial intelligence. For example, 'Numpy' for scientific computing, 'Pybrain' for machine learning, 'Scipy' for advanced computing and 'AIMA' for artificial intelligence.

##If-else Statement 
An if else Python statement evaluates whether an expression is true or false. If a condition is true, the “if” statement executes. Otherwise, the “else” statement executes. Python if else statements help coders control the flow of their programs.




###If statement

```python
num = 3
if num > 0:
    print(num, "is a positive number.")
print("This is always printed.")

num = -1
if num > 0:
    print(num, "is a positive number.")
print("This is also always printed.")
```

###If...else statement

```python
num = 3
if num >= 0:
    print("Positive or Zero")
else:
    print("Negative number")
```

###If...elif...else
The elif is short for else if. It allows us to check for multiple expressions.

If the condition for if is False, it checks the condition of the next elif block and so on.

If all the conditions are False, the body of else is executed.

Only one block among the several if...elif...else blocks is executed according to the condition.

The if block can have only one else block. But it can have multiple elif blocks.

```python
num = 3.4
if num > 0:
    print("Positive number")
elif num == 0:
    print("Zero")
else:
    print("Negative number")
```

##Loops
Something you must have noticed: our shows often show reruns. Python has a powerful concept, it makes use of so-called looping, which we can use to shorten repetitive code. 

###For loop
A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

This is less like the for keyword in other programming languages, and works more like an iterator method as found in other object-oriented programming languages.

With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc.


```python

numbers = [6, 5, 3, 8, 4, 2, 5, 4, 11]

# variable to store the sum
sum = 0

# iterate over the list
for val in numbers:
    sum = sum+val

print("The sum is", sum)
```

###While loop
In python, while loop is used to execute a block of statements repeatedly until a given a condition is satisfied. And when the condition becomes false, the line immediately after the loop in program is executed. 


```python
count = 0
while (count < 20):   
    count = count + 1
    print("Hello World")
```
